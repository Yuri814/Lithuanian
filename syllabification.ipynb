```python
import pandas as pd
import re

def syllabify(word):
    # Define Lithuanian vowels (including accented)
    vowels = "aeiouyąėįųų"

    # List of Lithuanian diphthongs (included accented)
    diphthongs = [
    "ai", "au", "ei", "eu", "ie", "iu", "ui", "uo", "ie",
    "ąi", "ąu", "ėi", "ėu", "įe", "įu", "ųi", "ųo"
    ]
    # List of common Lithuanian digraphs
    digraphs = ["ch", "cz", "dz", "dž", "j", "š", "ž"]

    # First, handle diphthongs by replacing them with a placeholder to ensure they stay together
    for diph in diphthongs:
        word = word.replace(diph, diph.upper())  # Convert diphthongs to uppercase for easy handling
    
    # Then, handle digraphs by replacing them with placeholders too
    for dig in digraphs:
        word = word.replace(dig, dig.upper())  # Convert digraphs to uppercase for easy handling

    # Regular expression to break word into syllables (handling vowels and consonants). Since `vowels` doesn't contain uppercases, it won't break diphthongs nor digraphs
    pattern = r'([^' + vowels + ']*[' + vowels + ']+[^' + vowels + ']*)'
    
    # Apply the pattern to break the word into syllables
    syllables = re.findall(pattern, word)

    # Join syllables and revert uppercase placeholders back to normal
    syllables = [syl.lower() for syl in syllables]

    return '-'.join(syllables)

```

```python
csv_file = "lithuanian_words_1page.csv"
df = pd.read_csv(csv_file)
df['syllables'] = df['Dark Text'].apply(syllabify)
```
