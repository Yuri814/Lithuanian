```python
import pyphen
def syllabify(word):

    dic = pyphen.Pyphen(lang='lt')
    syllables = dic.inserted(word).split('-')

    return syllables
```

```python
import pandas as pd
csv_file = "lithuanian_words_1page.csv"
df = pd.read_csv(csv_file)
df['syllables'] = df['Dark Text'].apply(syllabify)
```
